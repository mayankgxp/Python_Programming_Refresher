```python
import json as JSON 
```

```python
### declare global variables ###
expenses = []
monthly_budget = 0.0
```

```python
### function to add expense ###
def addExpense():
    print("Adding expense...")
    amount = float(input("Enter amount"))
    description = input("Enter description")
    category = input("Enter category (Food, Travel, Utility, etc.)")
    date = input("Enter date (YYYY-MM-DD)")

    expense = {
    "amount": amount,
     "description": description,
     "category":category,
     "date": date
    }
    expenses.append(expense)
```

```python
### function to set monthly budget ###
def setMonthlyBudget():
    monthly_budget = float(input("Set your monthly budget: "))
    print(f"Monthly budget set to {monthly_budget}")
```

```python
### function to view expenses ###   
def viewExpenses():
    for expense in expenses:
        print(expense)
```

```python
### function to save expenses to a file ###
def save_to_file():
    data = {
        "expenses": expenses,
        "monthly_budget": monthly_budget
    }

    with open("expenses.json", "w") as f:
      JSON.dump(data, f)
```

```python
### function to load expenses from a file ###
def load_from_file():
    with open("expenses.json", "r") as f:
        data = JSON.load(f)

    expenses = data.get("expenses", [])
    monthly_budget = data.get("monthly_budget", 0.0)
    print("expenses", expenses)
    print("monthly_budget", monthly_budget)
```

```python
### function to view budget status ###
def view_budget_status():
    total_expense = sum(expense["amount"] for expense in expenses)
    print(f"Total expenses: {total_expense}")
    remaining_budget = monthly_budget - total_expense
    if(remaining_budget <= 0):
        print("You have exceeded your budget!")
    else:
        print(f"Remaining budget: {remaining_budget}")
```

```python
### function to show menu ###
def show_menu():
    print("1. Add Expense")
    print("2. Set Monthly Budget")
    print("3. View Expenses")
    print("4. View Budget Status")
    print("5. Save Data")
    print("6. Load Data")
    print("7. Exit")
```

```python
### main function ###
def main():
    while True:
        show_menu()
        choice = input("Enter your choice:")
        if choice == '1':
            addExpense()
        elif choice == '2':
            setMonthlyBudget()
        elif choice == '3':
            viewExpenses()
        elif choice == '4':
            view_budget_status()
        elif choice == '5':
            save_to_file()
        elif choice == '6':
            load_from_file()
        elif choice == '7':
            break
        else:
            print("Invalid choice. Please try again.")
```

```python
main()
```
